```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_data_dir = 'Train'
test_data_dir = 'test'
```

```python
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(150, 150),  # Resize images to 150x150
    batch_size=32,
    class_mode='categorical'  
)

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(16, activation='softmax'))  
```

```python
model.summary()
```

```python
from keras.optimizers import RMSprop
optimizer = RMSprop(learning_rate=0.001)  # Learning rate set to 0.001
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=30,
    validation_data=test_generator,
    validation_steps=test_generator.samples // test_generator.batch_size
)

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(test_generator, verbose=2)
print(f'Test accuracy: {test_acc}')
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(range(10), acc, label='Training Accuracy')
plt.plot(range(10), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(10), loss, label='Training Loss')
plt.plot(range(10), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load the saved model
model = load_model('my_cnn_modell.h5')

# Assuming 'test_data_dir' is the path to your test data directory
test_data_dir = 'test'
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load your pre-trained model
model = load_model('my_cnn_modell.h5')

# Select a sample image for demonstration
img_path = 'test/Aphids/zoom_1.jpg'
img = image.load_img(img_path, target_size=(150, 150))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)

# Predict class probabilities
preds = model.predict(x)
class_idx = np.argmax(preds[0])
class_output = model.output[:, class_idx]

# Get the last convolutional layer in your model
last_conv_layer = model.get_layer('conv2d_11')  # Replace 'last_conv_layer_name' with the name of your last convolutional layer

# Compute gradients of the top predicted class with respect to the output feature map
with tf.GradientTape(persistent=True) as tape:
    preds = model.predict(x)
    top_pred_index = np.argmax(preds[0])
    top_class_channel = model.output[:, top_pred_index]

# Extract gradients
grads = tape.gradient(top_class_channel, last_conv_layer.output)

# Pool gradients across each feature map channel
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

# Multiply each channel in the feature map array by "how important this channel is" with regard to the top predicted class
last_conv_layer_output = last_conv_layer.output[0]
heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
heatmap = tf.squeeze(heatmap)

# Normalize the heatmap
heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)

# Plot the heatmap
plt.matshow(heatmap)
plt.show()

# Superimpose the heatmap on the original image
img = image.load_img(img_path, target_size=(150, 150))
img = image.img_to_array(img)

heatmap = tf.image.resize(heatmap, (img.shape[1], img.shape[0]))

heatmap = tf.keras.preprocessing.image.array_to_img(heatmap)
heatmap = heatmap.convert("RGB")

heatmap = image.img_to_array(heatmap)

superimposed_img = heatmap * 0.4 + img
superimposed_img /= np.max(superimposed_img)

plt.imshow(superimposed_img)
plt.axis('off')
plt.show()

# Release the tape resources
del tape
```
